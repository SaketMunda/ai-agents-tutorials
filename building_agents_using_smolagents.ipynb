## Selecting a playlist for the party using `smolagents`

We can build an agent capable of searching the web. To give the agent access to this tool, we include it in the tool list when creating the agent.

For the model, we'll rely on `InferenceModel`, which provided access to the Hugging Face's [InferenceAPI](https://huggingface.co/docs/api-inference/index). The default model is `Qwen/Qwen2.5-Coder-32B-Instruct`, which is performant and available for fast inference, but this is optional we can chooose model of our own.

Running an agent is quite straightforward.

```python
from smolagents import InferenceClientModel, CodeAgent, WebSearchTool

agent = CodeAgent(
    tools=[WebSearchTool()],
    model = InferenceClientModel())

agent.run("Search for the best music recommendations for a party at the Wayne's mansion.")
```

## Using a Custom Tool to Prepare the Menu

Now that we have selected a playlist, we need to organise the menu for the guests. Again, Alfred can take advantage of `smolagents` to do so.

Here, we use the `@tool` decorator to define a custom function that acts as a tool.

For now we will build rule-based tool to just use that custom tool inside our `CodeAgent`.

```python
from smolagents import CodeAgent, tool

@tool
def suggest_menu(occasion: str) -> str:
  """
  Suggests a menu based on the occasion.
  Args:
    occasion (str): The type of occasion for the party. Allowed values are:
                    - "casual": Menu for casual party.
                    - "formal": Menu for formal party.
                    - "superhero": Menu for superhero party.
                    - "custom": Custom Menu
  """
  if occasion == "casual":
    return "Pizza, snack, and drinks."
  elif occasion == "formal":
    return "3-course dinner with wine and dessert."
  elif occasion == "superhero":
    return "Buffet with high-energy and healthy food."
  else:
    return "Custom menu for the butler."

agent = CodeAgent(
    tools=[suggest_menu],
    model=InferenceClientModel("Qwen/Qwen2.5-Coder-32B-Instruct"),)

agent.run("Prepare a formal menu for the party.")
```

## Using Python imports inside the Agent

We have the playlist and the menu ready, but we need to check one more crucial detail: preparation time!

Alfred needs to calculate when everything would be ready if he started preparing now, in case they need assistance from other superheroes.

`smolagents` specialized in agents that write and execute Python code snippets, offering sandboxed execution for security. It supports both open-source and proprietary language models, making it adaptable to various development environments.

**Code execution has strict security measures** - imports outside a predefined safe list are blocked by default. However, you can authorize additional imports by passing them as strings in `additional_authorized_imports`. For more details on secure code execution, see the official [guide](https://huggingface.co/docs/smolagents/tutorials/secure_code_execution).

```python
from smolagents import InferenceClientModel, CodeAgent
import numpy as np
import time
import datetime

agent = CodeAgent(
    tools=[],
    model=InferenceClientModel("Qwen/Qwen2.5-Coder-32B-Instruct"),
    additional_authorized_imports=['datetime']
)


agent.run(
    """
    Alfred needs to prepare for the party. Here are the tasks:
    1. Prepare the drinks - 30 minutes.
    2. Decorate the mansion - 60 minutes
    3. Set up the menu - 45 minutes
    4. Prepare the music and playlist - 45 minutes

    If we start right now, at what time will the party be ready?
    """
)
```
